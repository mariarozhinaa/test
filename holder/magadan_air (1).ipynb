```python
data_t = data[['Date', 'T']]
data_t
```

```python
data_t['Date'] = pd.to_datetime(data_t['Date']).dt.date
```

```python
data_pre = data[['Date', 'RRR', 'sss']]
```

```python
data_pre['Date'] = data_pre['Date'].dt.to_period('M')
```

```python
data_pre['Date']
```

```python
data_pre['RRR'].unique()
```

```python
data_pre['sss'].unique()
```

```python
data_pre = data_pre.query('RRR != ["Следы осадков", "Осадков нет"] & sss != ["Снежный покров не постоянный.", "Менее 0.5", \
                "Измерение невозможно или неточно."]')
```

```python
data_pre = data_pre.groupby('Date')[['RRR', 'sss']].agg({'RRR':'mean','sss':'mean'}).reset_index()
```
